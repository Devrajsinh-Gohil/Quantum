```python
from qiskit import *
import math
from qiskit.visualization import plot_bloch_multivector
```

```python
qc = QuantumCircuit(1)
qc.draw(output='mpl')
```

```python
sim = Aer.get_backend('statevector_simulator')
state  = execute(qc,backend=sim).result().get_statevector()
plot_bloch_multivector(state)
```

```python
qc.x(0)
qobj = assemble(qc)
state  = execute(qc,backend=sim).result().get_statevector()
plot_bloch_multivector(state)
```

```python
qc.reset(0)
qc.h(0)
state  = execute(qc,backend=sim).result().get_statevector()
plot_bloch_multivector(state)
```

```python
qc.z(0)
state  = execute(qc,backend=sim).result().get_statevector()
plot_bloch_multivector(state)
```

```python
qc.reset(0)
qc.h(0)
qc.rz(math.pi/4,0)
state  = execute(qc,backend=sim).result().get_statevector()
plot_bloch_multivector(state)
```

```python
qc.ry(math.pi/4,0)
state  = execute(qc,backend=sim).result().get_statevector()
plot_bloch_multivector(state)
```

```python
qc.reset(0)
qc.h(0)
qc.s(0)
state  = execute(qc,backend=sim).result().get_statevector()
plot_bloch_multivector(state)
```

```python
qc.reset(0)
qc.h(0)
qc.t(0)
state  = execute(qc,backend=sim).result().get_statevector()
plot_bloch_multivector(state)
```
